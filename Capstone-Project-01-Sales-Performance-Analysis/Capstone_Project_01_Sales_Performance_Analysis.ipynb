```python
import pandas as pd

# Load Dataset
df = pd.read_csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
print("========== BUSINESS OBJECTIVE ==========\n")

print("""
The objective of this project is to analyze the company's sales
performance using historical sales data.

The analysis focuses on:

1. Sales Trend Analysis
2. Region-wise Sales Performance
3. Growth & Decline Insights
4. Business Recommendations

The goal is to identify opportunities for improving business performance
through data-driven decisions.
""")
```

```python
print(df.columns.tolist())
```

```python
print("========== SALES SUMMARY ==========\n")

print(f"Total Sales     : ${df['Sales'].sum():,.2f}")
print(f"Total Profit    : ${df['Profit'].sum():,.2f}")
print(f"Total Quantity  : {df['Quantity'].sum()}")
print(f"Average Discount: {df['Discount'].mean()*100:.2f}%")
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print("========== CATEGORY-WISE SALES ==========\n")
print(category_sales)

plt.figure(figsize=(7,5))
category_sales.plot(kind="bar")

plt.title("Category-wise Sales Performance")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

highest_region = region_sales.idxmax()
lowest_region = region_sales.idxmin()

highest_sales = region_sales.max()
lowest_sales = region_sales.min()

print("========== GROWTH & DECLINE INSIGHTS ==========\n")

print(f"Highest Performing Region : {highest_region} (${highest_sales:,.2f})")
print(f"Lowest Performing Region  : {lowest_region} (${lowest_sales:,.2f})")

print("\nInsights:")
print(f"- {highest_region} contributes the highest sales and represents the strongest business region.")
print(f"- {lowest_region} records the lowest sales and needs performance improvement.")
```

```python
print("========== BUSINESS RECOMMENDATIONS ==========\n")

print("1. Increase investment in the West region to maximize sales.")
print("2. Improve sales performance in lower-performing regions.")
print("3. Focus more on Technology products as they generate the highest sales.")
print("4. Optimize discount strategies to improve profitability.")
print("5. Continue promoting high-performing product categories.")
print("6. Use sales insights to support future business decisions.")

print("\n========== CONCLUSION ==========\n")

print("The sales performance analysis shows that business growth is primarily driven by high-performing regions and product categories. By strengthening weaker regions and optimizing business strategies, the company can improve overall sales and profitability.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
print("========== BUSINESS OBJECTIVE ==========\n")

print("""
The objective of this project is to analyze customer purchasing behavior
using the Sample Superstore dataset.

The analysis focuses on:

1. Customer Segmentation
2. Purchase Patterns
3. Retention & Churn Insights
4. Business Recommendations

The goal is to understand customer behavior and support better
business decision-making through data analysis.
""")
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

print("========== CUSTOMER SEGMENTATION ==========\n")
print(segment_sales)

plt.figure(figsize=(7,5))
segment_sales.plot(kind="bar")

plt.title("Customer Segmentation by Sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
category_quantity = df.groupby("Category")["Quantity"].sum().sort_values(ascending=False)

print("========== PURCHASE PATTERNS ==========\n")
print(category_quantity)

plt.figure(figsize=(7,5))
category_quantity.plot(kind="bar")

plt.title("Purchase Pattern by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Quantity Purchased")

plt.tight_layout()
plt.show()
```

```python
segment_orders = df.groupby("Segment").size().sort_values(ascending=False)

print("========== RETENTION & CHURN INSIGHTS ==========\n")
print(segment_orders)

highest_segment = segment_orders.idxmax()
lowest_segment = segment_orders.idxmin()

print(f"\nHighest Purchasing Segment : {highest_segment}")
print(f"Lowest Purchasing Segment  : {lowest_segment}")

print("\nInsights:")
print(f"- {highest_segment} customers contribute the highest number of purchases and are more likely to be retained.")
print(f"- {lowest_segment} customers show comparatively lower purchase activity and may require targeted marketing to improve retention.")
```

```python
print("========== BUSINESS RECOMMENDATIONS ==========\n")

print("1. Focus more on the Consumer segment as it contributes the highest sales.")
print("2. Create personalized offers for Corporate and Home Office customers.")
print("3. Promote high-demand product categories to increase purchase volume.")
print("4. Use targeted marketing campaigns to engage lower-performing customer segments.")
print("5. Monitor customer purchasing behavior regularly to improve retention.")

print("\n========== CONCLUSION ==========\n")

print("The customer behavior analysis shows that different customer segments contribute differently to overall sales. Understanding purchase patterns and focusing on customer engagement strategies can improve customer retention, increase sales, and support long-term business growth.")
```

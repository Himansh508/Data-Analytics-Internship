```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
print("========== DATASET INFORMATION ==========")
print(df.info())

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

print("\n========== DUPLICATE RECORDS ==========")
print(df.duplicated().sum())
```

```python
print("========== SUMMARY STATISTICS ==========")

display(df.describe())
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

print("========== REGION WISE SALES ==========")
print(region_sales)
```

```python
plt.figure(figsize=(8,5))

region_sales.plot(kind="bar")

plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print("========== CATEGORY WISE SALES ==========")
print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

print("========== REGION WISE PROFIT ==========")
print(region_profit)
```

```python
region_profit.plot(kind="bar")

plt.title("Region Wise Profit")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.show()
```

```python
subcategory_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

print("========== TOP 10 SUB-CATEGORIES BY SALES ==========")
print(subcategory_sales)
```

```python
subcategory_sales.plot(kind="bar")

plt.title("Top 10 Sub-Category Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

print("========== SEGMENT WISE SALES ==========")
print(segment_sales)
```

```python
segment_sales.plot(kind="bar")

plt.title("Segment Wise Sales")
plt.xlabel("Segment")
plt.ylabel("Total Sales")

plt.show()
```

```python
profit_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

print("========== CATEGORY WISE PROFIT ==========")
print(profit_category)
```

```python
profit_category.plot(kind="bar")

plt.title("Category Wise Profit")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.show()
```

```python
import pandas as pd

# Load Dataset
df = pd.read_csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
print("========== BUSINESS KPIs ==========\n")

# Overall KPIs
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = len(df)
total_quantity = df["Quantity"].sum()
avg_discount = df["Discount"].mean() * 100
profit_margin = (total_profit / total_sales) * 100

print(f"Total Sales       : ${total_sales:,.2f}")
print(f"Total Profit      : ${total_profit:,.2f}")
print(f"Total Orders      : {total_orders}")
print(f"Total Quantity    : {total_quantity}")
print(f"Average Discount  : {avg_discount:.2f}%")
print(f"Profit Margin     : {profit_margin:.2f}%")
```

```python
region_report = df.groupby("Region")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False)

print("========== REGION PERFORMANCE ==========\n")
print(region_report)
```

```python
category_report = df.groupby("Category")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False)

print("========== CATEGORY PERFORMANCE ==========\n")
print(category_report)
```

```python
subcategory_profit = (
    df.groupby("Sub-Category")[["Profit"]]
      .sum()
      .sort_values(by="Profit", ascending=False)
      .head(10)
)

print("========== TOP 10 PROFITABLE SUB-CATEGORIES ==========\n")
print(subcategory_profit)
```

```python
print("\n========== EXECUTIVE BUSINESS SUMMARY ==========\n")

top_region = df.groupby("Region")["Sales"].sum().idxmax()
top_category = df.groupby("Category")["Sales"].sum().idxmax()
top_segment = df.groupby("Segment")["Sales"].sum().idxmax()
top_subcategory = df.groupby("Sub-Category")["Profit"].sum().idxmax()

print(f"Top Performing Region      : {top_region}")
print(f"Top Performing Category    : {top_category}")
print(f"Top Customer Segment       : {top_segment}")
print(f"Most Profitable Sub-Category : {top_subcategory}")

print("\nBusiness Recommendations")
print("- Increase investment in the highest-performing region.")
print("- Focus more on Technology products to maximize revenue.")
print("- Improve sales strategy in lower-performing regions.")
print("- Strengthen marketing for Corporate and Home Office customers.")
print("- Continue promoting high-profit product categories.")
```

```python
import pandas as pd

# Load Dataset
df = pd.read_csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
print("========== BUSINESS PROBLEM ==========\n")

print("""
ShopSphere wants to analyze its sales performance to identify
business growth opportunities.

The management wants answers to the following questions:

1. Which region generates the highest sales?
2. Which product category performs the best?
3. Which customer segment contributes the most revenue?
4. Which products generate the highest profit?
5. What business decisions can improve overall performance?
""")
```

```python
print("========== BUSINESS KPIs ==========\n")

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = len(df)
total_quantity = df["Quantity"].sum()
avg_discount = df["Discount"].mean() * 100
profit_margin = (total_profit / total_sales) * 100

print(f"Total Sales       : ${total_sales:,.2f}")
print(f"Total Profit      : ${total_profit:,.2f}")
print(f"Total Orders      : {total_orders}")
print(f"Total Quantity    : {total_quantity}")
print(f"Average Discount  : {avg_discount:.2f}%")
print(f"Profit Margin     : {profit_margin:.2f}%")
```

```python
import matplotlib.pyplot as plt

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
segment_sales.plot(kind="bar")

plt.title("Sales by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
top_products = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(9,5))
top_products.plot(kind="bar")

plt.title("Top 10 Most Profitable Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")

plt.tight_layout()
plt.show()
```

```python
print("========== FINAL BUSINESS INSIGHTS ==========\n")

top_region = df.groupby("Region")["Sales"].sum().idxmax()
top_category = df.groupby("Category")["Sales"].sum().idxmax()
top_segment = df.groupby("Segment")["Sales"].sum().idxmax()
top_subcategory = df.groupby("Sub-Category")["Profit"].sum().idxmax()

print(f"Top Performing Region      : {top_region}")
print(f"Top Performing Category    : {top_category}")
print(f"Top Customer Segment       : {top_segment}")
print(f"Most Profitable Sub-Category : {top_subcategory}")

print("\n========== BUSINESS RECOMMENDATIONS ==========\n")

print("1. Increase investment in the West region to maximize sales.")
print("2. Focus on Technology products for higher revenue growth.")
print("3. Improve performance in lower-performing regions through targeted marketing.")
print("4. Strengthen relationships with Corporate and Home Office customers.")
print("5. Continue promoting high-profit sub-categories such as Copiers.")
print("6. Monitor discounts carefully to protect profit margins.")

print("\n========== CONCLUSION ==========\n")
print("The analysis shows that the West region, Technology category, Consumer segment, and Copiers sub-category are the key business drivers. Strategic investments and better marketing can further improve the company's overall performance.")
```
